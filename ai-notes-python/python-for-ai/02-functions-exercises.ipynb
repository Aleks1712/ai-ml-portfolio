## Combining Conditional Statements and Functions

Define a function, called **compare_the_two()**, with two arguments. If the first one is greater than the second one, let it print "Greater". If the second one is greater, it should print "Less". Let it print "Equal" if the two values are the same number.

```python
def compare_the_two(x, y):
    if x > y:
        print("Greater")
    elif x < y:
        print("Less")
    else:
        print("Equal")


compare_the_two(10, 10)
compare_the_two(10, 5)
compare_the_two(5, 10)
```

Create a function, called "distance_from_zero", that returns the absolute value of a provided single argument and prints a statement "Not Possible" if the argument provided is not a number.
Call the funtion with the values of -10 and "cat" to verify it works correctly.

```python
def distance_from_zero(x):
    if isinstance(x, (int, float)):
        return abs(x)

    print("Not Possible")
    return None


distance_from_zero(-10)
distance_from_zero("cat")
```

Skriv `clean_text(text)` som:

- `strip()`
- `lower()`
- bytter flere spaces til én space

Test med f.eks. `"  Hello   AI  "`.

## B) Batching (ML)

Skriv `make_batches(xs, batch_size)` som returnerer en liste av batches.

- `xs=[1,2,3,4,5]`, `batch_size=2` → `[[1,2],[3,4],[5]]`

## C) Sikker lookup i dict

Skriv `safe_get(d, key, default=None)` som bruker `dict.get` og fungerer for alle typer.


```python
import re
from typing import Any, Dict, List, Sequence, TypeVar

T = TypeVar("T")


def clean_text(text: str) -> str:
    """Strip + lowercase + collapse multiple spaces."""
    text = text.strip().lower()
    text = re.sub(r"\s+", " ", text)
    return text


def make_batches(xs: Sequence[T], batch_size: int) -> List[List[T]]:
    if batch_size <= 0:
        raise ValueError("batch_size must be > 0")
    return [list(xs[i : i + batch_size]) for i in range(0, len(xs), batch_size)]


def safe_get(d: Dict[Any, Any], key: Any, default=None):
    return d.get(key, default)


print(clean_text("  Hello   AI  "))
print(make_batches([1, 2, 3, 4, 5], 2))
print(safe_get({"a": 1}, "missing", default=0))
```
