## Combining Conditional Statements and Functions

*Suggested Answers follow (usually there are multiple ways to solve a problem in Python).*

Define a function, called **compare_the_two()**, with two arguments. If the first one is greater than the second one, let it print "Greater". If the second one is greater, it should print "Less". Let it print "Equal" if the two values are the same number.

```python
def compare_the_two(x,y):
    if x > y:
        print ("Greater")
    elif x < y:
        print ("Less")
    else:
        print ("Equal")
```

```python
compare_the_two(10,10)
```

Create a function, called "distance_from_zero", that returns the absolute value of a provided single argument and prints a statement "Not Possible" if the argument provided is not a number.
Call the funtion with the values of -10 and "cat" to verify it works correctly.

```python
def distance_from_zero(x):
    if type(x) == int or type(x) == float:
        return abs(x)
    else:
        print ("Not possible")
```

# Utvidelser (bedre praksis)

## 1) Litt bedre type-sjekk enn `type(x) == int`

I stedet for:

- `type(x) == int or type(x) == float`

bruk heller:

- `isinstance(x, (int, float))`

Det håndterer også subklasser.

## 2) `raise` vs `print("Not possible")`

I prosjekter vil du ofte:

- `raise TypeError(...)` (tydelig feil)

Men i kursoppgaver kan `print` være ok.

## 3) Navn og konsistens

Jeg anbefaler å unngå skrivefeil i funksjonsnavn (f.eks. `divisin_by_2`). Det funker, men det blir vanskelig å lese/vedlikeholde.


```python
from __future__ import annotations


def distance_from_zero(x):
    """Return abs(x) if x is a number; otherwise print message.

    (Kept in the spirit of the original exercise.)
    """

    if isinstance(x, (int, float)):
        return abs(x)

    print("Not possible")
    return None


distance_from_zero(-10), distance_from_zero("cat")
```
